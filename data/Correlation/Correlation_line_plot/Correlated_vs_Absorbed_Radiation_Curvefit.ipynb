### This code plot the most pearson's correlated Biochemical processes against the amount of cosmic radiations absorbed in-flights, post-flight and a combination of both. 

```python
import pandas as pd
from scipy.stats import pearsonr
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('../../All_strain.xlsx')
df
```

### Inflight vs Post-flight

```python
# In-flight
di = df[df['Return_Days'] < 1]
di
```

```python
# Post-flight
dp = df[df['Return_Days'] > 0]
dp
```

### All Samples

```python
# Curve fit of most correlated Mitochondria Processes against amount of radiation absorbed

# Prepare x data
x = df["mGy_Absorbed"]

# Dictionary of targets to compare
targets = {
    "MT_Central_Dogma": "red",
    "OXPHOS": "green",
    "MT_Ribosome": "blue",
    "MT_Protein_Import": "magenta",
    "Complex_I_V": "orange", 
}

    
# Plot setup
plt.figure(figsize=(9, 6))
sns.set(style="white")

# Plot each target variable
for label, color in targets.items():
    y = df[label]
    if len(df) >= 3:
        z = np.polyfit(x, y, 2)
        p = np.poly1d(z)
        x_vals = np.linspace(min(x), max(x), 200)
        plt.plot(x_vals, p(x_vals), label=label, color=color)
        # plt.plot(x_vals, p(x_vals), label=label, color=color)

# Add horizontal line at y = 0
plt.axhline(y=0, color='black', linewidth=1.5, linestyle='--')

# Customize plot

plt.xlabel("Amount Radiation Absorbed (mGy)", fontsize=16)
plt.ylabel("Normalized Enrichment Scores", fontsize=16)
plt.legend(title="Mitochondrial Processes")
plt.tick_params(labelsize=14, pad=1)

# Customize spines
ax = plt.gca()
ax.spines["top"].set_linewidth(3)
ax.spines["left"].set_linewidth(3)
ax.spines["right"].set_linewidth(3)
ax.spines["bottom"].set_linewidth(3)

plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

plt.tight_layout()
# plt.savefig('Correlation/Most_Correlated_Mitochondrial_processes_Vs_Radiation_Absorbed.svg',format='svg', dpi=600, transparent=False)
plt.show()
```

```python
# Prepare x data
x = di["mGy_Absorbed"]
xp = dp["mGy_Absorbed"]

# Define groups of targets
group_1 = {
    "MT_Central_Dogma": "red",
    "OXPHOS": "green",
    "MT_Ribosome": "blue",
}

group_2 = {
    "MT_Protein_Import": "magenta",
    "Complex_I_V": "orange",
}

# Plot setup
fig, axes = plt.subplots(1, 2, figsize=(18, 7.5), sharey=True)
sns.set(style="white")

# Helper function to plot each group
def plot_group(ax, group, title):
    for label, color in group.items():
        if label not in di.columns or label not in dp.columns:
            continue  # skip if missing
        y = di[label]
        if len(di) >= 3:
            z = np.polyfit(x, y, 2)
            p = np.poly1d(z)
            x_vals = np.linspace(min(x), max(x), 200)
            ax.plot(x_vals, p(x_vals), label=f"{label}_IF", color=color)

        yp = dp[label]
        if len(dp) >= 3:
            zp = np.polyfit(xp, yp, 2)
            pp = np.poly1d(zp)
            xp_vals = np.linspace(min(xp), max(xp), 200)
            ax.plot(xp_vals, pp(xp_vals), label=f"{label}_PF", color=color, linestyle='--')

    ax.axhline(y=0, color='black', linewidth=1.5, linestyle='--')
    # ax.set_title(title, fontsize=16, weight='bold')
    ax.set_xlabel("Radiation Absorbed (mGy)", fontsize=17, weight='bold')
    ax.tick_params(labelsize=14)
    ax.spines["top"].set_linewidth(2)
    ax.spines["left"].set_linewidth(2)
    ax.spines["right"].set_linewidth(2)
    ax.spines["bottom"].set_linewidth(2)
    ax.legend(title="Mitochondrial Processes", fontsize=11)

# Plot both subplots
plot_group(axes[0], group_1, "Group 1: Central Dogma, OXPHOS, Ribosome")
plot_group(axes[1], group_2, "Group 2: Protein Import, Complex I–V")

# Shared y-label
fig.text(0.08, 0.5, 'Normalized Enrichment Scores', va='center', rotation='vertical', fontsize=18, weight='bold')
# plt.tight_layout()
plt.subplots_adjust(wspace=0.06)
plt.savefig('Most_Correlated_Mitochondrial_processes_Vs_Radiation_Absorbed_IF&PF.svg',format='svg', dpi=600, transparent=False)
plt.show()
```
