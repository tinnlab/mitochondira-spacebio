### This code computes the pearson's correlation for mitochondrial processes in samples exposed to cosmic radiations in-flights, post-flight and a combination of both. Bonferroni corrections were applied to identify false positives.

```python
import pandas as pd
from scipy.stats import pearsonr
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('../../All_strain.xlsx')
df
```

### Inflight vs Post-flight

```python
# In-flight
di = df[df['Return_Days'] < 1]
di
```

```python
# Post-flight
dp = df[df['Return_Days'] > 0]
dp
```

#### All data

```python
### Pearson correlation of Mitochondria process against the amount of amount of radiation absorbed
### Correlation for all samples

ref_col_name = df.columns[2]  # Column 3 - amount of radiation absorbed (mGy_Absorbed)
target_cols = df.columns[9:21]  # Columns 10 to 21

results = []

# Step 1: Pearson correlations
for col in target_cols:
    x = df[ref_col_name]
    y = df[col]

    valid = x.notna() & y.notna()
    x_valid = x[valid]
    y_valid = y[valid]

    if len(x_valid) >= 2:
        r, p = pearsonr(x_valid, y_valid)
        results.append({'Processes': col, 'correlation': r, 'p_value': p})
    else:
        results.append({'Processes': col, 'correlation': None, 'p_value': None})

results_df = pd.DataFrame(results)

# Step 2: Adjust p-values (Bonferroni correction)
results_df = results_df.dropna(subset=['p_value'])
adjusted = multipletests(results_df['p_value'], method='bonferroni')
results_df['p_value_adj'] = adjusted[1]
results_df['significant'] = adjusted[0]

# Step 3: Sort by absolute correlation
results_df['abs_corr'] = results_df['correlation'].abs()
results_df_sorted = results_df.sort_values(by='abs_corr', ascending=False)
results_df_sorted
results_df_sorted.to_csv('Correlation_Mitochondria_vs_Radiation_AllSample.csv', index=False)


# Step 4: Plot top N correlations
top_n = 20
top_results = results_df_sorted.head(top_n)
```

```python

# plt.figure(figsize=(10, 6))
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(top_results['processes'], top_results['correlation'], color=['green' if s else 'gray' for s in top_results['significant']])
plt.xlabel("Pearson Correlation with Amount Radiation Absorbed", fontsize = 13.0)

plt.axvline(x=0, color='black', linewidth=0.8)
plt.tick_params(labelsize=11, pad=1)

ax.spines["top"].set_linewidth(3)
ax.spines["left"].set_linewidth(3)
ax.spines["right"].set_linewidth(3)
ax.spines["bottom"].set_linewidth(3)

plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

plt.gca().invert_yaxis()  # highest on top
plt.tight_layout()
plt.savefig('Correlation_Mitochondria_vs_Radiation_AllSample.tiff',format='tiff', dpi=300, transparent=False)
plt.show()
```
